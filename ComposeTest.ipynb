```julia
using Compose
using Color
```

```julia
compose(context(), fill("tomato"),
(context(0.3, 0.0, 0.3, 0.3),  circle(),(context(), text(0.5,0.5,"mami"), fill("black")) ),
        (context(0.6, 0.6, 0.3, 0.3), circle()),
        (context(0, 0.6, 0.3, 0.3), circle()))
```

```julia
function sierpinski(n)
    if n == 0
        compose(context(), polygon([(1,1), (0,1), (1/2, 0)]))
    else
        t = sierpinski(n - 1)
        compose(context(),
                (context(1/4,   0, 1/2, 1/2), t),
                (context(  0, 1/2, 1/2, 1/2), t),
                (context(1/2, 1/2, 1/2, 1/2), t))
    end
end
```

```julia
compose(sierpinski(6), linewidth(0.1mm), fill(nothing), stroke("black"))
```
